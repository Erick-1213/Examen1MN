```python
import math

def calculate_sqrt_root(n:float):
    for _ in range(5):
        raiz = round(math.sqrt(n), 6)
        n = raiz
    return n
```

```python
calculate_sqrt_root(2.0)
```

```python
def calulate_log_briggs(n: float):
    log_brigs = 32*(1/math.log(10))*(calculate_sqrt_root(n)-1)
    return log_brigs, abs( math.log10(n)-log_brigs)/math.log10(n)
```

```python
calulate_log_briggs(2.0)
```

```python
for i in range(2,11):
    print(i, calulate_log_briggs(float(i)))
```
