```python
import numpy as np
from scipy.optimize import bisect
```

```python
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625
```

```python
def encontrar_raices(f, inicio=-10, fin=10, paso=0.01, tol=1e-10):
    raices = []
    x_valores = np.arange(inicio, fin, paso)

    for i in range(len(x_valores) - 1):
        a = x_valores[i]
        b = x_valores[i + 1]

        fa = f(a)
        fb = f(b)
        if abs(fa) < tol:
            raiz = round(a, 10)

            if not any(abs(raiz - r) < 1e-6 for r in raices):
                raices.append(raiz)

        if fa * fb < 0:
            raiz = bisect(f, a, b, xtol=tol)
            raiz = round(raiz, 10)

            if not any(abs(raiz - r) < 1e-6 for r in raices):
                raices.append(raiz)

    raices.sort()
    return raices
```

```python
raices = encontrar_raices(f)
```

```python
print("Raices encontradas:")
for r in raices:
    print(r)
```
